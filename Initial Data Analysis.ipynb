Amount of profanity - most common terms

```python
import pandas as pd

data = pd.read_pickle('./pickle/dtm.pkl')
data = data.transpose()
data.head()
```

```python
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(30) #sort values in desc order and then choose top 30
    top_dict[c]= list(zip(top.index, top.values)) #key: index of business, value: words in reviews

top_dict
```

```python
for index, top_words in top_dict.items():
    print(index)
    print(', '.join([word for word, count in top_words[0:14]])) #just print the top 15 words for each business
    print('---')
```

```python
#some can be added to a stop words list
from collections import Counter

# Let's first pull out the top 30 words for each business
words = []
for index in data.columns:
    top = [word for (word, count) in top_dict[index]]
    for t in top:
        words.append(t)
        
words
```

```python
Counter(words).most_common()
```

```python
add_stop_words = [word for word, count in Counter(words).most_common() if count > 1000]
add_stop_words
```

```python
from sklearn.feature_extraction import text 
from sklearn.feature_extraction.text import CountVectorizer

# Read in cleaned data
cleaned = pd.read_pickle('./pickle/corpus.pkl')

# Add new stop words
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

# Recreate document-term matrix
cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(cleaned.text)
data_stop = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_stop.index = cleaned.index

#pickle for later use
```

### Number of words

```python
unique_list = []
full_names = cleaned.business_id

for index in data.columns:
    uniques = data[index].to_numpy().nonzero()[0].size
    unique_list.append(uniques)

# Create a new dataframe that contains this unique word count
data_words = pd.DataFrame(list(zip(full_names, unique_list)), columns=['business_id', 'unique_words'])
data_unique_sort = data_words.sort_values(by='unique_words')
data_unique_sort
```

```python
total_list = []
for index in data.columns:
    totals = sum(data[index])
    total_list.append(totals)

# Let's add some columns to our dataframe
data_words['total_words'] = total_list
data_words
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud


wc = WordCloud(stopwords=stop_words, background_color="white", colormap="Dark2",
               max_font_size=150, random_state=42).generate(cleaned.text[25])

plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import numpy as np

y_pos = np.arange(len(data_words))

plt.subplot(1, 2, 1)
plt.barh(y_pos, data_unique_sort.unique_words, align='center')
plt.yticks(y_pos, data_unique_sort.index)
plt.title('Number of Unique Words', fontsize=20)

plt.tight_layout()
plt.show()
```

### Amount of generic positive words (excellent, great, good, etc)

```python
Counter(words).most_common()
```

```python
data_good_words = data.transpose()[['excellent', 'great', 'good', 'awesome']]
data_positives = pd.concat([data_good_words.excellent + data_good_words.awesome, data_good_words.great + data_good_words.good], axis=1)
data_positives.columns = ['excellent/awesome', 'great/good']
data_positives
```

```python
plt.rcParams['figure.figsize'] = [10, 8]

for i, businessIndex in enumerate(data_positives.index):
    x = data_positives['excellent/awesome'].loc[businessIndex]
    y = data_positives['great/good'].loc[businessIndex]
    plt.scatter(x, y, color='blue')
    plt.text(x+1.5, y+0.5, full_names[i], fontsize=10)
    plt.xlim(0, 15) 
    
plt.title('Number of Positive Words Used in Review', fontsize=20)
plt.xlabel('Number of Excellents/Awesomes', fontsize=15)
plt.ylabel('Number of Goods/Greats', fontsize=15)

plt.show()
```
