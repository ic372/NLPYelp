Use Markov chains to generate text! We assume that the next word is only dependent on the previous word.

```python
import pandas as pd

data = pd.read_pickle('./pickle/corpus.pkl')
data
```

```python
business1 = data.text[0]
business1[:200] #just use the first business
```

Build a Markov Chain that creates a dictionary, where keys are all words in corpus, and values are list of words that follow the keys


```python
from collections import defaultdict

def markov_chain(text):
    words = text.split(' ')
    m_dict = defaultdict(list) #initialize a dictionary with a list as the values
    for current_word, next_word in zip(words[0:-1], words[1:]):
        m_dict[current_word].append(next_word)
    m_dict = dict(m_dict)
    return m_dict
```

```python
business1_dict = markov_chain(business1)
business1_dict
```

Make a generator to create our synthetic text! Should take in dictionary above and number of words to generate.

```python
import random

def generate_sentence(chain, count=20):
    word1 = random.choice(list(chain.keys()))
    sentence = word1.capitalize()
    for i in range(count-1):
        word2 = random.choice(chain[word1])
        word1 = word2
        sentence += ' ' + word2
    sentence += '.'
    return(sentence)
```

```python
generate_sentence(business1_dict)
```
